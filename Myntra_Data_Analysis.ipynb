##### 3.8) Finding the Brand with most number of Products 

```python
# how many products are there for greenfibre brand name
data=[]
for brand_name in df['brand_name'].unique():
  c=0;
  for i in df['brand_name']:
    if i==brand_name:
      c+=1

  data.append([brand_name ,c])
data=pd.DataFrame(data,columns=['name','product'])

data.sort_values(by='product',ascending=False)
```

##### 3.9) Finding the top 10 most expensive Brands

```python
#Doing above question similarly by using Group by clause
df.groupby('brand_name')# it will collect unique brand name and makes a group
df.groupby('brand_name').size().sort_values(ascending=False)# it will ckeck how many products are there in each brand name 
```

```python
# to find the expensive brand we will find the avg price of a product in perticular brand
df.groupby('brand_name').mean()['marked_price'].sort_values(ascending=False)
```

```python
#brands which are giving discounts is
df.groupby('brand_name').mean()['discount_percent'].sort_values(ascending=False)
```
